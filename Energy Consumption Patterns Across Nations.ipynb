```python
# Calculate null percentages
total_rows=3649
null_columns = df.columns[df.isnull().any()]
null_percentages = df[null_columns].isnull().sum() / total_rows * 100

# Bar chart
plt.figure(figsize=(14, 8))
null_percentages.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Percentage of Null Values in Columns', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=12)
plt.xlabel('Columns', fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Calculate the percentage of null values in each column
null_percentages = (df.isnull().sum() / len(df)) * 100

# Drop columns where the null percentage is greater than 50%
df_cleaned = df.dropna(axis=1, thresh=len(df) * 0.5)

# Display the cleaned DataFrame
print("Columns retained after dropping those with more than 50% null values:")
print(df_cleaned.columns)
```

```python
# Drop the 'Year' column
#df_cleaned = df_cleaned.drop(columns=['Year'])


# Grouping by 'Entity' and calculating the mean for each entity
df_avg = df.drop(columns=['Year']).groupby('Entity').mean().reset_index()

# Display the resulting dataframe
df_avg.head()


```
